In the convolutional layer has 2 batches/filters, each is 3*3 pixels,and each has a single input channel



Input Tensor: <br>
1 axes(-): number of images in the batch, the number could very with batch size
number of color channels <br>
2, 3, 4: image height, width, number of color channels<br>
Convolutional layer:<br>

1, 2 axes: height, weight of filter batch <br>
3 axes: must has the same size as the number of color channels in the input images <br>
4 axes: must be equal to the number of output channels we want to obtain after the convolution, this is also the number of filters we are going to learn in the layer

Stride: the number of pixels used to shift convolution
